```python
import pandas as pd
users = pd.read_table(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')
useroccupation = users.set_index(["occupation"])
# step 4 mean age per occupation
meanage = useroccupation["age"].groupby('occupation').mean()
meanage
```

```python
# step 5 
gen_occup = users.groupby('gender')['occupation'].value_counts()
sex_ratio = gen_occup['M']/gen_occup['F']
print(sex_ratio.sort_values(ascending=False))
```

```python
# step 6 
maximum_age_ = useroccupation["age"].groupby('occupation').max()
minimum_age_=  useroccupation["age"].groupby('occupation').min()
print(maximum_age_, minimum_age_)
```

```python
# step 7 
users.groupby(['occupation','gender'])['age'].mean()
```

```python
 #step 8 
gen_occup = users.groupby('gender')['occupation'].value_counts()
sexratio_male = gen_occup['M']/(gen_occup['M'] + gen_occup['F'])*100
sexratio_female = gen_occup['F']/(gen_occup['M'] + gen_occup['F'])*100
print(sexratio_male, sexratio_female)
```

# Question 4


Wind Statistics

```python
#1 Import the necessary libraries
import pandas as pd
```

```python
#2 Import the dataset from the attached file wind.txt
data = pd.read_csv('wind.txt', delimiter='\s+')
```

```python
#3 Assign it to a variable called data and replace the first 3 columns by a proper datetime index.
data.rename(columns={"Yr":"Year", "Mo":"Month", "Dy":"Day"}, inplace=True)
data
```

```python
#4 Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
def fixYear():
    data['Year'] += 1900 #

fixYear() # Fix year by converting it in 4-digit value

data
```

```python
#5 Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
data.index = pd.to_datetime(pd.concat([data['Year'], data['Month'], data['Day']], axis=1)) # setting the index

data.drop(['Year','Month','Day'], axis=1, inplace=True) # dropping first 3 columns Year, Month & Day

data
```

```python
#6 Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.
data.isna().sum() 
```

```python
#7 Compute how many non-missing values there are in total
data.notnull().sum().sum()
```

```python
#8 Calculate the mean windspeeds of the windspeeds over all the locations and all the times. A single number for the entire dataset.
data.mean().mean()
```

```python
#9 Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days. A different set of numbers for each location.

loc_stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 
loc_stats.rename(columns={0:'Min', 1:'Max', 2:'Mean', 3:'Std'}, inplace=True) 
loc_stats
```

```python
#10 Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.

day_stats = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)
day_stats.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)
day_stats
```

```python
#11 Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.
avg_windspeed_jan = data[data.index.month == 1].mean()

avg_windspeed_jan
```

```python
#12 Downsample the record to a yearly frequency for each location.
yearlyfreq = data.resample('Y').ffill()

yearlyfreq
```

```python
#13 Downsample the record to a monthly frequency for each location.
monthlyfreq = data.resample('M').ffill()

monthlyfreq
```

```python
#14 Downsample the record to a weekly frequency for each location.
weeklyfreq = data.resample('W').ffill()

weeklyfreq
```

```python
#15 Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.
weekly_stats = weeklyfreq[:52]

weekly_stats = pd.concat([weekly_stats.mean(), weekly_stats.max(), weekly_stats.mean(), weekly_stats.std()], axis=1)
weekly_stats.rename(columns={0:'Mean',1:'Max',2:'Mean',3:'Std'},inplace=True)

weekly_stats
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

roman_emperor_reigns = pd.read_csv('roman-emperor-reigns.csv')
cause_Ofdeath = roman_emperor_reigns.groupby('Cause_of_Death').count()[['Emperor']]
cause_Ofdeath
emperor = pd.Series(cause_Ofdeath['Emperor'])
cause_death = pd.Series(cause_Ofdeath.index)
explode = (0.1,0,0,0,0,0,0,0)
fig2, ax2 = plt.subplots()
ax2.pie(emperor, explode=explode, labels=cause_death,autopct='%1.1f%%', shadow=False, startangle=90)
ax2.axis('equal')
plt.title('Cause of Death of Roman Emperors')
plt.show()
```
